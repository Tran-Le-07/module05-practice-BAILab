```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/tranle/Desktop/iris_EDA/iris.csv")
```

```python
df.info()
```

```python
clean_df = df.dropna()
```

```python
desc = clean_df.describe().T
print("\nDescriptive statistics:")
display(desc)
```

```python
species_col = "variety"
sepal_length_col = "sepal.length"
petal_length_col = "petal.length"
```

```python
plt.figure(figsize=(6, 5))
for species, group in clean_df.groupby(species_col):
    plt.scatter(
        group[sepal_length_col],
        group[petal_length_col],
        label=species,
        alpha=0.7
    )
plt.xlabel(sepal_length_col)
plt.ylabel(petal_length_col)
plt.title("Sepal length vs Petal length by species")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
